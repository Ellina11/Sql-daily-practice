```python
# Create a DataFrame manually
data = [
    (1, "Ali Hassan",    "Engineering", 9000),
    (2, "Siti Rahimah",  "Engineering", 8500),
    (3, "Kumar Selvam",  "Engineering", 8500),
    (4, "John Lim",      "Engineering", 7000),
    (5, "Nurul Ain",     "Marketing",   7500),
    (6, "Wei Ling",      "Marketing",   7500),
    (7, "Priya Nair",    "Marketing",   6800),
    (8, "Farid Azman",   "HR",          6500),
    (9, "Mei Shan",      "HR",          6200),
    (10, "Rajan Pillai", "HR",          6200),
]

columns = ["employee_id", "name", "department", "salary"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
# Select specific columns
df.select("name", "department", "salary").show()

# Filter rows
df.filter(df.salary > 7000).show()
```

```python
from pyspark.sql.functions import when, col

# Add salary_grade column
df2 = df.withColumn("salary_grade",
    when(col("salary") >= 8000, "High")
    .when(col("salary") >= 7000, "Mid")
    .otherwise("Low")
)
df2.show()
```

```python
from pyspark.sql.functions import avg, max, count, round

# Average salary per department
df.groupBy("department").agg(
    round(avg("salary"), 2).alias("avg_salary"),
    max("salary").alias("max_salary"),
    count("employee_id").alias("headcount")
).show()
```

```python
# Create a departments DataFrame
dept_data = [
    ("Engineering", "Kuala Lumpur"),
    ("Marketing",   "Petaling Jaya"),
    ("HR",          "Cyberjaya"),
]

dept_df = spark.createDataFrame(dept_data, ["department", "location"])

# Inner join
joined_df = df.join(dept_df, on="department", how="inner")
joined_df.select("name", "department", "salary", "location").show()
```

```python
# orderBy — sort by salary descending
df.orderBy(col("salary").desc()).show()
```

```python
# Write as Delta table (better than Parquet in Databricks)
df2.write.mode("overwrite").saveAsTable("employees_clean")

# Read it back to verify
df_check = spark.read.table("employees_clean")
df_check.show()
print(f"Rows loaded back: {df_check.count()}")
```

```python
# Recreate DataFrame
data = [
    (1, "Ali Hassan",    "Engineering", 9000),
    (2, "Siti Rahimah",  "Engineering", 8500),
    (3, "Kumar Selvam",  "Engineering", 8500),
    (4, "John Lim",      "Engineering", 7000),
    (5, "Nurul Ain",     "Marketing",   7500),
    (6, "Wei Ling",      "Marketing",   7500),
    (7, "Priya Nair",    "Marketing",   6800),
    (8, "Farid Azman",   "HR",          6500),
    (9, "Mei Shan",      "HR",          6200),
    (10, "Rajan Pillai", "HR",          6200),
]

columns = ["employee_id", "name", "department", "salary"]
df = spark.createDataFrame(data, columns)
print("DataFrame ready!")
```

```python
# Register DataFrame as a temp view
df.createOrReplaceTempView("employees")

# Now query it with SQL
spark.sql("""
    SELECT department,
           ROUND(AVG(salary), 2) AS avg_salary,
           COUNT(*) AS headcount
    FROM employees
    GROUP BY department
    ORDER BY avg_salary DESC
""").show()
```

```python
from pyspark.sql.functions import dense_rank, lag, col
from pyspark.sql.window import Window

# Define window spec
window_dept = Window.partitionBy("department").orderBy(col("salary").desc())

# Add rank column
df_ranked = df.withColumn("rnk", dense_rank().over(window_dept))
df_ranked.show()
```

```python
from pyspark.sql.functions import lag

# LAG — previous salary within department
window_dept = Window.partitionBy("department").orderBy(col("salary").desc())

df_lag = df.withColumn("prev_salary", lag("salary", 1).over(window_dept))
df_lag.show()
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

# Define the function
def salary_band(salary):
    if salary >= 8000:
        return "Senior"
    elif salary >= 7000:
        return "Mid"
    else:
        return "Junior"

# Register as UDF
salary_band_udf = udf(salary_band, StringType())

# Apply to DataFrame
df_band = df.withColumn("band", salary_band_udf(col("salary")))
df_band.show()
```

```python
from pyspark.sql.functions import broadcast

# Small lookup table — department locations
dept_data = [
    ("Engineering", "Kuala Lumpur"),
    ("Marketing",   "Petaling Jaya"),
    ("HR",          "Cyberjaya"),
]
dept_df = spark.createDataFrame(dept_data, ["department", "location"])

# Broadcast the small table
df_joined = df.join(broadcast(dept_df), on="department", how="inner")
df_joined.select("name", "department", "salary", "location").show()
```

```python
from pyspark.sql.functions import coalesce, lit

# Simulate nulls
data_with_nulls = [
    (1, "Ali Hassan",    "Engineering", 9000),
    (2, "Siti Rahimah",  None,          8500),
    (3, "Kumar Selvam",  "Engineering", None),
    (4, None,            "HR",          6500),
]
cols = ["employee_id", "name", "department", "salary"]
df_nulls = spark.createDataFrame(data_with_nulls, cols)

# Show nulls
df_nulls.show()

# Drop rows with ANY null
df_nulls.dropna().show()

# Fill nulls with defaults
df_nulls.fillna({
    "department": "Unknown",
    "salary": 0,
    "name": "Unknown"
}).show()

# coalesce — use first non-null value
df_nulls.withColumn("dept_clean",
    coalesce(col("department"), lit("Unknown"))
).show()
```

```python
# Raw data — simulating source system extract
raw_data = [
    (1, "Ali Hassan",    "Engineering", 9000, "2021-03-15"),
    (2, "Siti Rahimah",  "Engineering", 8500, "2020-07-01"),
    (3, "Kumar Selvam",  "Engineering", 8500, "2022-01-10"),
    (4, "John Lim",      "Engineering", 7000, "2023-06-20"),
    (5, "Nurul Ain",     "Marketing",   7500, "2021-09-05"),
    (6, "Wei Ling",      "Marketing",   7500, "2020-11-30"),
    (7, "Priya Nair",    "Marketing",   6800, "2022-04-18"),
    (8, "Farid Azman",   "HR",          6500, "2021-01-25"),
    (9, "Mei Shan",      "HR",          6200, "2023-03-08"),
    (10, "Rajan Pillai", "HR",          6200, "2020-08-14"),
    (10, "Rajan Pillai", "HR",          6200, "2020-08-14"),  # duplicate
    (11, None,           "HR",          6000, "2023-01-01"),  # null name
]

cols = ["employee_id", "name", "department", "salary", "hire_date"]
bronze_df = spark.createDataFrame(raw_data, cols)

# Write to Bronze Delta table — no transformation, raw as-is
bronze_df.write.mode("overwrite").saveAsTable("bronze_employees")

print(f"Bronze layer: {bronze_df.count()} rows loaded")
bronze_df.show()
```

```python
from pyspark.sql.functions import to_date

# Read from Bronze
bronze = spark.read.table("bronze_employees")

# Clean
silver_df = bronze \
    .dropDuplicates() \
    .dropna(subset=["name"]) \
    .withColumn("hire_date", to_date(col("hire_date"), "yyyy-MM-dd")) \
    .withColumn("salary", col("salary").cast("integer"))

# Write to Silver Delta table
silver_df.write.mode("overwrite").saveAsTable("silver_employees")

print(f"Silver layer: {silver_df.count()} rows after cleaning")
silver_df.show()
```

```python
# Update gold table — simulate a new run with different data
gold_df_v2 = silver.groupBy("department").agg(
    round(avg("salary") * 1.1, 2).alias("avg_salary"),  # 10% salary increase
    max("salary").alias("max_salary"),
    min("salary").alias("min_salary"),
    count("employee_id").alias("headcount")
).orderBy("avg_salary", ascending=False)

gold_df_v2.write.mode("overwrite").saveAsTable("gold_dept_summary")

print("Version 2 — after 10% salary increase:")
spark.read.table("gold_dept_summary").show()

# Query previous version using time travel
print("Version 1 — original:")
spark.sql("SELECT * FROM gold_dept_summary VERSION AS OF 0").show()
```

```python
from delta.tables import DeltaTable

# New incoming data — some updates, some new
new_data = [
    (1, "Ali Hassan",    "Engineering", 9500, "2021-03-15"),  # update salary
    (11, "Zara Ahmad",   "Marketing",   7200, "2024-01-15"),  # new employee
]
new_df = spark.createDataFrame(new_data,
    ["employee_id", "name", "department", "salary", "hire_date"])

# Get reference to Silver Delta table
silver_table = DeltaTable.forName(spark, "silver_employees")

# MERGE
silver_table.alias("target").merge(
    new_df.alias("source"),
    "target.employee_id = source.employee_id"
).whenMatchedUpdateAll() \
 .whenNotMatchedInsertAll() \
 .execute()

print("After MERGE:")
spark.read.table("silver_employees").orderBy("employee_id").show()
```

```python
# OPTIMIZE — compacts small files into larger ones for better performance
spark.sql("OPTIMIZE silver_employees")
print("OPTIMIZE done!")

# Check table history
spark.sql("DESCRIBE HISTORY silver_employees").select(
    "version", "timestamp", "operation"
).show(truncate=False)
```
