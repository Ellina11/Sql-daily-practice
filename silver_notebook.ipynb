```python
from pyspark.sql.functions import col, to_date

# Read from Bronze
bronze = spark.read.table("bronze_employees")

# Clean
silver_df = bronze \
    .dropDuplicates() \
    .dropna(subset=["name"]) \
    .withColumn("hire_date", to_date(col("hire_date"), "yyyy-MM-dd")) \
    .withColumn("salary", col("salary").cast("integer"))

# Write to Silver
silver_df.write.mode("overwrite").saveAsTable("silver_employees")
print(f"Silver done: {silver_df.count()} rows")
```
