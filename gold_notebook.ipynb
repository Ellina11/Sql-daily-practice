```python
from pyspark.sql.functions import avg, max, min, count, round

# Read from Silver
silver = spark.read.table("silver_employees")

# Aggregate
gold_df = silver.groupBy("department").agg(
    round(avg("salary"), 2).alias("avg_salary"),
    max("salary").alias("max_salary"),
    min("salary").alias("min_salary"),
    count("employee_id").alias("headcount")
).orderBy("avg_salary", ascending=False)

# Write to Gold
gold_df.write.mode("overwrite").saveAsTable("gold_dept_summary")
print(f"Gold done: {gold_df.count()} rows")
gold_df.show()
```
